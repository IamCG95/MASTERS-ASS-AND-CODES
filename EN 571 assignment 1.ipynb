```python
import matplotlib.pyplot as plt
import numpy as np

# Define the constraints
x = np.linspace(0, 200, 400)

# Constraint 1: 2x + y <= 400 (as equality for plotting)
y_constraint1 = 400 - 2*x

# Constraint 2: x <= 150
y_constraint2 = np.full_like(x, 200)

# Constraint 3: y <= 200
x_constraint3 = np.full_like(x, 150)

# Iso-profit lines for Z = 2000, 2400, 2800, and 4000
iso_profit_2000 = (2000 - 8*x) / 5
iso_profit_2400 = (2400 - 8*x) / 5
iso_profit_2800 = (2800 - 8*x) / 5
iso_profit_4000 = (4000 - 8*x) / 5

# Plotting the constraints
plt.figure(figsize=(10, 8))

plt.plot(x, y_constraint1, label=r'$2x + y = 400$', color='blue')
plt.axvline(x=150, label=r'$x = 150$', color='green', linestyle='--')
plt.axhline(y=200, label=r'$y = 200$', color='orange', linestyle='--')

# Plotting the iso-profit lines
plt.plot(x, iso_profit_2000, label=r'$Z = 2000$', color='red', linestyle=':')
plt.plot(x, iso_profit_2400, label=r'$Z = 2400$', color='red', linestyle=':')
plt.plot(x, iso_profit_2800, label=r'$Z = 2800$', color='red', linestyle=':')
plt.plot(x, iso_profit_4000, label=r'$Z = 4000$', color='purple', linestyle=':')

# Feasible region shading
plt.fill_between(x, np.minimum(y_constraint1, 200), where=(x <= 150), color='gray', alpha=0.3)

# Setting the plot limits
plt.xlim(0, 200)
plt.ylim(0, 250)

# Labels and title
plt.xlabel(r'$x$ (Type 1 Hats)')
plt.ylabel(r'$y$ (Type 2 Hats)')
plt.title('The Graphical Solution of the Linear Programming Problem')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
import itertools

# Job data: (processing time, due date, late penalty)
jobs = {
    1: {'processing_time': 5, 'due_date': 25, 'late_penalty': 19},
    2: {'processing_time': 20, 'due_date': 22, 'late_penalty': 12},
    3: {'processing_time': 15, 'due_date': 35, 'late_penalty': 34}
}

# Calculate the total penalty for a given job sequence
def calculate_penalty(sequence):
    total_time = 0
    total_penalty = 0
    for job in sequence:
        total_time += jobs[job]['processing_time']
        if total_time > jobs[job]['due_date']:
            total_penalty += (total_time - jobs[job]['due_date']) * jobs[job]['late_penalty']
    return total_penalty

# Generate all possible sequences of jobs
job_ids = list(jobs.keys())
sequences = list(itertools.permutations(job_ids))

# Find the sequence with the minimum total penalty
best_sequence = None
min_penalty = float('inf')

for sequence in sequences:
    penalty = calculate_penalty(sequence)
    if penalty < min_penalty:
        min_penalty = penalty
        best_sequence = sequence

best_sequence, min_penalty
```

```python
from pulp import LpProblem, LpMinimize, LpVariable, lpSum, LpBinary, LpStatus, PULP_CBC_CMD

# Define the data for processing time, due date, and late penalty for each job
processing_time = {1: 5, 2: 20, 3: 15}
due_date = {1: 25, 2: 22, 3: 35}
late_penalty = {1: 19, 2: 12, 3: 34}

# Create the linear programming problem
model = LpProblem("Job_Sequencing_Min_Late_Penalty", LpMinimize)

# Decision variables: x[i, j] indicates if job j is in position i
x = LpVariable.dicts("x", ((i, j) for i in range(1, 4) for j in range(1, 4)), cat=LpBinary)
# Completion time of job j
C = LpVariable.dicts("C", range(1, 4), lowBound=0)
# Lateness of job j
late = LpVariable.dicts("late", range(1, 4), lowBound=0)

# Objective function: Minimize the total late penalty
model += lpSum(late_penalty[j] * late[j] for j in range(1, 4))

# Constraints

# 1. Each job should occupy exactly one position
for j in range(1, 4):
    model += lpSum(x[i, j] for i in range(1, 4)) == 1

# 2. Each position should have exactly one job assigned
for i in range(1, 4):
    model += lpSum(x[i, j] for j in range(1, 4)) == 1

# 3. Define completion time constraints based on the sequence and calculate lateness
for j in range(1, 4):
    # Completion time constraint
    model += C[j] >= lpSum(x[i, j] * sum(processing_time[k] for k in range(1, i+1)) for i in range(1, 4))
    # Lateness constraint
    model += late[j] >= C[j] - due_date[j]
    model += late[j] >= 0

# Solve the model
model.solve(PULP_CBC_CMD(msg=True))

# Output results
if LpStatus[model.status] == "Optimal":
    print("Optimal Job Sequence:")
    job_sequence = sorted(((i, j) for i in range(1, 4) for j in range(1, 4) if x[i, j].varValue == 1), key=lambda t: t[0])
    for i, j in job_sequence:
        print(f"Position {i}: Job {j}")
    
    total_penalty = sum(late_penalty[j] * late[j].varValue for j in range(1, 4))
    print(f"Total Late Penalty: {total_penalty}")
else:
    print("No optimal solution found.")
```

```python
from pulp import LpMinimize, LpProblem, LpVariable, lpSum

# Data from the problem
processing_times = [5, 20, 15]
due_dates = [25, 22, 35]
late_penalties = [19, 12, 34]

# Number of jobs
n_jobs = len(processing_times)

# Create the LP problem
model = LpProblem("Job_Sequencing", LpMinimize)

# Decision variables
# x[i][j] is 1 if job j is assigned to position i, 0 otherwise
x = [[LpVariable(f"x_{i}_{j}", cat="Binary") for j in range(n_jobs)] for i in range(n_jobs)]

# Auxiliary variables for late penalties
# L[j] is the lateness of job j
L = [LpVariable(f"L_{j}", lowBound=0) for j in range(n_jobs)]

# Objective function: Minimize the sum of weighted lateness penalties
model += lpSum(late_penalties[j] * L[j] for j in range(n_jobs))

# Constraints

# 1. Each position can only have one job
for i in range(n_jobs):
    model += lpSum(x[i][j] for j in range(n_jobs)) == 1

# 2. Each job must be assigned to one position
for j in range(n_jobs):
    model += lpSum(x[i][j] for i in range(n_jobs)) == 1

# 3. Lateness constraints for each job based on due date and position
for j in range(n_jobs):
    model += L[j] >= lpSum((lpSum(processing_times[k] * x[k][j] for k in range(i + 1)) - due_dates[j]) * x[i][j] 
                           for i in range(n_jobs))

# Solve the problem
model.solve()

# Extract the results
job_sequence = [[int(x[i][j].value()) for j in range(n_jobs)] for i in range(n_jobs)]
late_penalty_values = [L[j].value() for j in range(n_jobs)]
objective_value = model.objective.value()

print("Job Sequence:", job_sequence)
print("Late Penalty Values:", late_penalty_values)
print("Objective Value (Minimum Penalty):", objective_value)
```
