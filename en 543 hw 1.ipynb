```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
u = 16.5 ### lambad avlue form poisson
k = 20
prob_poisson = stats.poisson.cdf(k,u)
print(f"The probability of finding 20 or fewer defects is approximately: {prob_poisson:.4f}")
```

```python


# Define the rate parameter (lambda)
mu = 16.5

# Calculate the probability mass function (PMF) at k = 3
k = 20
prob = stats.poisson.pmf(k, mu)
print(f"Probability of {k} events: {prob}")

# Calculate the cumulative distribution function (CDF) at k = 3
cdf = stats.poisson.cdf(k, mu)
print(f"Probability of {k} or fewer events: {cdf}")

```
